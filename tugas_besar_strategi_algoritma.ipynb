```python
# QuickSort
def partition(arr, low, high):
    # Menentukan elemen pivot
    pivot = arr[high]
    # Mengatur indeks untuk elemen yang lebih kecil dari pivot
    i = low - 1
    # Iterasi melalui elemen dalam rentang low hingga high-1
    for j in range(low, high):
        # Jika elemen saat ini lebih kecil dari pivot
        if arr[j] < pivot:
            # Tingkatkan indeks elemen yang lebih kecil
            i += 1
            # Tukar elemen di indeks i dan j
            swap(arr, i, j)
    # Tukar elemen pivot ke posisi yang benar
    swap(arr, i + 1, high)
    # Kembalikan indeks partisi
    return i + 1

# Fungsi untuk menukar dua elemen dalam array
def swap(arr, low, high):
    arr[low], arr[high] = arr[high], arr[low]

# Fungsi utama untuk algoritma quicksort
def quicksort(arr, low, high):
    # Pastikan rentang array valid untuk diurutkan
    if low < high:
        # Dapatkan indeks partisi
        pi = partition(arr, low, high)
        # Rekursif mengurutkan bagian kiri dari partisi
        quicksort(arr, low, pi - 1)
        # Rekursif mengurutkan bagian kanan dari partisi
        quicksort(arr, pi + 1, high)
```

```python
# MergeSort

def merge(arr, left, mid, right):
    # Hitung ukuran array kiri dan kanan
    n1 = mid - left + 1
    n2 = right - mid
    # Buat array sementara untuk menyimpan data
    left_array = [arr[left + i] for i in range(n1)]
    right_array = [arr[mid + 1 + j] for j in range(n2)]
    # Inisialisasi indeks untuk iterasi pada kedua array
    i = j = 0
    k = left
    # Gabungkan array sementara ke array utama
    while i < n1 and j < n2:
        # Pilih elemen yang lebih kecil dan tambahkan ke array utama
        if left_array[i] <= right_array[j]:
            arr[k] = left_array[i]
            i += 1
        else:
            arr[k] = right_array[j]
            j += 1
        k += 1
    # Salin elemen yang tersisa dari array kiri, jika ada
    while i < n1:
        arr[k] = left_array[i]
        i += 1
        k += 1
    # Salin elemen yang tersisa dari array kanan, jika ada
    while j < n2:
        arr[k] = right_array[j]
        j += 1
        k += 1

# Fungsi utama untuk algoritma MergeSort
def merge_sort(arr, left, right):
    # Pastikan rentang valid untuk dibagi lebih lanjut
    if left < right:
        # Hitung titik tengah
        mid = (left + right) // 2
        # Rekursif membagi bagian kiri
        merge_sort(arr, left, mid)
        # Rekursif membagi bagian kanan
        merge_sort(arr, mid + 1, right)
        # Gabungkan hasil bagian kiri dan kanan
        merge(arr, left, mid, right)
```

```python
# SelectionSort
def selection_sort(arr):
    # Hitung panjang array
    n = len(arr)
    # Iterasi untuk setiap elemen dalam array
    for i in range(n - 1):
        # Anggap elemen saat ini sebagai yang terkecil
        min_idx = i
        # Cari elemen terkecil di sisa array
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        # Tukar elemen terkecil dengan elemen saat ini
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
```

```python
import time
import pandas as pd
from tabulate import tabulate
import matplotlib.pyplot as plt

# Example of selecting a column to sort by (e.g., math_score)
dataset = dataframe['math_score'].tolist()
print(dataset)  # Display the math scores as a list

def compare_sorting_algorithms(dataset):
    input_sizes = [10, 100, 500, 1000, 2000]
    algorithms = {
        "QuickSort": lambda arr: quicksort(arr, 0, len(arr) - 1),
        "MergeSort": lambda arr: merge_sort(arr, 0, len(arr) - 1),
        "SelectionSort": selection_sort
    }
    results = []

    for size in input_sizes:
        print(f"\nTesting for input size: {size}")
        data = dataset[:size]  # Use subset of the dataset
        row = {"Input Size": size}  # Initialize row with input size

        # Create a DataFrame for displaying the unsorted data in a table
        unsorted_df = pd.DataFrame(data, columns=["Original Data"])
        print(f"\nOriginal Data for Input Size {size}:")
        print(tabulate(unsorted_df, headers='keys', tablefmt='pretty', showindex=False))

        for name, func in algorithms.items():
            arr_copy = data[:]  # Copy array to ensure same input for all algorithms
            print(f"\nRunning {name}...", end=" ")
            start_time = time.time()
            if name == "SelectionSort" and size > 2000:
                print("skipped (too slow)")
                row[name] = None
                continue
            func(arr_copy)
            elapsed_time = time.time() - start_time
            row[name] = elapsed_time
            print(f"done in {elapsed_time:.4f} seconds")

            # Create a DataFrame for displaying the sorted data in a table
            sorted_df = pd.DataFrame(arr_copy, columns=[f"Sorted by {name}"])
            print(f"\nSorted Data by {name} for Input Size {size}:")
            print(tabulate(sorted_df, headers='keys', tablefmt='pretty', showindex=False))

        results.append(row)

    # Convert results into a DataFrame for better visualization of sorting times
    df = pd.DataFrame(results)

    # Display the comparison table of sorting times using tabulate
    print("\nSorting Algorithms Comparison (in Seconds):")
    print(tabulate(df, headers='keys', tablefmt='pretty', showindex=False))

    # Plot results for performance comparison
    plt.figure(figsize=(10, 6))
    for name in algorithms.keys():
        plt.plot(df['Input Size'], df[name], label=name, marker='o')

    plt.title("Comparison of Sorting Algorithms")
    plt.xlabel("Input Size (n)")
    plt.ylabel("Running Time (seconds)")
    plt.yscale("log")  # Log scale for better visualization
    plt.legend()
    plt.grid(True)
    plt.show()

# Run the comparison
compare_sorting_algorithms(dataset)
```
